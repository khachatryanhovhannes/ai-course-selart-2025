#### Step 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Step 2

```python
df = pd.read_csv("../data/car_data.csv")
```

#### Step 3

```python
# First 5 rows
print(df.head())
```

```python
# Info summary
print(df.info())
```

```python
# Statistical summary
print(df.describe())
```

#### Step 4

```python
# Calculate the average price of all cars
average_price = df["price_usd"].mean()

print(f"The average car price is: ${average_price:,.2f}")
```

#### Step 5

```python
# Create a boolean mask: True for rows where horsepower is greater than 300
mask = df["hp"] > 300

# Use boolean indexing to filter the original DataFrame
powerful_cars = df[mask].copy()

# reset the index so it starts from 0 in the new DataFrame
powerful_cars.reset_index(drop=True, inplace=True)

# Print the resulting DataFrame
print(powerful_cars)
```

#### Step 6

```python
import matplotlib.pyplot as plt

# Create a new figure with a defined size
plt.figure(figsize=(10, 6))

# Plot a bar chart: model names on the x-axis, prices on the y-axis
plt.bar(df["model"], df["price_usd"])

# Add a title and labels to make the chart more informative
plt.title("Price of Car Models")
plt.xlabel("Car Model")
plt.ylabel("Price (USD)")

# Rotate x-axis labels so they don't overlap
plt.xticks(rotation=45)

# Adjust the layout so labels and titles fit well
plt.tight_layout()

# Save the figure as a PNG file
plt.savefig("../figures/price_bar_chart.png")

# Display the chart in the notebook
plt.show()
```

#### Step 7

```python
import matplotlib.pyplot as plt

# Create a new figure for the scatter plot
plt.figure(figsize=(8, 5))

# Scatter plot: x-axis = horsepower, y-axis = price
plt.scatter(df["hp"], df["price_usd"])

# Add title and labels
plt.title("Horsepower vs. Price")
plt.xlabel("Horsepower")
plt.ylabel("Price (USD)")

# Add grid lines for easier reading
plt.grid(True)

# Adjust layout to fit everything nicely
plt.tight_layout()

# Save the chart as PNG (optional)
plt.savefig("../figures/hp_vs_price_scatter.png")

# Show the plot
plt.show()
```
